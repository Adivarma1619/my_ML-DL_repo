```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
id = pd.read_csv('iris.csv')
id.head()
```

```python
id.describe()
```

```python
id.info()
```

```python
colors = ['red', 'green', 'blue']
species = ['Iris-setosa ', 'Iris-versicolor', 'Iris-virginica']
```

```python
iris_dummies = pd.get_dummies(id['species'], prefix='species')*1
iris_dummies
```

```python
id_m = pd.concat([id, iris_dummies], axis=1).drop(['species','species_Iris-virginica'],axis=1)
id_m
```

```python
x = id_m.iloc[:,0:4].values
y = id_m.iloc[:,[4,5]].values[:,0]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

```

```python

cm = confusion_matrix(model.predict(X_test),y_test)  
cm
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Predicted')
plt.xlabel('Actual')
plt.show()
```
